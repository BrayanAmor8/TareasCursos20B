```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos vectores LI

#Oblicua
def vectores_base(alpha,phi):

    a = np.array([1,0])
    b1 = alpha * np.linalg.norm(a)**2*np.cos(phi)/a[0]
    b2 = np.sqrt((alpha*a[0])**2-b1**2)
    b = np.array([b1,b2])
    return (a,b)
```

```python
#Create a list of vectors
def list_vec(n,alpha,phi):
    vector = vectores_base(alpha,phi)
    vectors = []
    for i in range(-n,n):
        for j in range(n):
            vectors.append(i*vector[0] + j*vector[1])
    return vectors
```

```python
#Plot the array of vectors
def graph(n,alpha, phi):
    vector = vectores_base(alpha,phi)
    vectors = list_vec(n,alpha, phi)
    origin = np.array([0,0])
    plt.figure(figsize=(4,4))
    plt.grid(alpha=0.5, linestyle="--")
    plt.scatter([v[0] for v in vectors], [v[1] for v in vectors], s=70, alpha =0.7)
    plt.xlim(-1.5,6)
    plt.ylim(-0.5,6)
    plt.quiver(*origin, vector[0][0], vector[0][1], scale_units="xy", scale=1)
    plt.quiver(*origin, vector[1][0], vector[1][1], scale_units="xy", scale=1)
    plt.show
```

```python
graph(20, 2, np.pi/6)
graph(20, 1.2, np.pi/3)
graph(20, 1, 2*np.pi/3)
```
